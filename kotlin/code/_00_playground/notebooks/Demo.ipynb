# Imports

```kotlin
import lang.Objects.DiceRoller

val rolls = List(10_000) { DiceRoller.roll(20) }
rolls
```

```kotlin
rolls.average()
```

# Play with Charts

```kotlin
%use kandy
```

```kotlin
val faceFrequency = rolls.groupingBy { it }.eachCount()
faceFrequency
```

```kotlin
plot {
    bars {
        x(faceFrequency.keys)
        y(faceFrequency.values)
    }
}
```

```kotlin
val moreRolls = List(10_000) { DiceRoller.rollWithAdvantage(20)}
val moreFaceFrequency = moreRolls.groupingBy { it }.eachCount()
plot {
    bars {
        x(moreFaceFrequency.keys)
        y(moreFaceFrequency.values)
    }
}
```

# We can add extension functions
Disadvantage: to roll with disadvantage means to take the lower of two dice throws

```kotlin
fun DiceRoller.rollWithDisadvantage(sides: Int): Int =
    min(roll(sides), roll(sides))
```

```kotlin
val lessRolls = List(10_000) { DiceRoller.rollWithDisadvantage(20)}
val lessFaceFrequency = lessRolls.groupingBy { it }.eachCount()
plot {
    bars {
        x(lessFaceFrequency.keys)
        y(lessFaceFrequency.values)
    }
}
```
